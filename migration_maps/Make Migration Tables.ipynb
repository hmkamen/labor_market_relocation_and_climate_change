```python
import pandas as pd
import numpy as np
import scipy.stats
import math
import warnings
warnings.filterwarnings("ignore")
```

```python
####create unskilled migration Tables
sl=pd.read_csv('/Users/hannahkamen/Downloads/flowmap_location_lookup.csv')
all_unskl_entrance=pd.read_csv('/Users/hannahkamen/Downloads/unskilled_entrance3.csv')
all_unskl_entrance['count']=all_unskl_entrance['count']*14412444.2

all_unskl_entrance=all_unskl_entrance.merge(sl,left_on='origin', right_on='id')
all_unskl_entrance=all_unskl_entrance.merge(sl,left_on='dest', right_on='id')
all_unskl_entrance=all_unskl_entrance.rename(columns={'name_x':'1','name_y':'2'})
```

```python
#####get unskilled entrance table
all_unskl_entrance_lm=all_unskl_entrance[all_unskl_entrance['1']!=all_unskl_entrance['2']]
unskl_sorted=all_unskl_entrance_lm.sort_values(by='count',ascending=False)[['1','2','count']]
unskl_sorted=unskl_sorted.rename(columns={'1':'Origin','2':'Destination','count':'Number of Migrants'})
####sumby destination state
top_dest=unskl_sorted.groupby('Destination',as_index=False).sum().sort_values(by='Number of Migrants',ascending=False).head(10)['Destination']
#####
#limit to top 10 dest
unskl_sorted_lm=unskl_sorted[unskl_sorted['Destination'].isin(top_dest)]

unskl_sorted_lm=unskl_sorted_lm.groupby('Destination',as_index=False).agg({'Number of Migrants':max})
unskl_sorted_lm=unskl_sorted_lm.merge(unskl_sorted[['Origin','Number of Migrants']],on='Number of Migrants')

unskl_sorted_lm['Number of Migrants']=unskl_sorted_lm['Number of Migrants'].astype(int).apply(lambda x: '{:,}'.format(x))
unskl_sorted_lm=unskl_sorted_lm[['Origin','Destination','Number of Migrants']]
###print to latex
print(unskl_sorted_lm.to_latex(index=False, column_format='rrrr'))
```
